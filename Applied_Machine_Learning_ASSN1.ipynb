## Part 1
Recreating a matplotlib graph picked from: https://www.tylervigen.com/spurious-correlations

I chose to recreate "Computer science doctorates awarded in the US correlates with Total revenue generated by arcades".

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

years = [
  2000,
  2001,
  2002,
  2003,
  2004,
  2005,
  2006,
  2007,
  2008,
  2009,
]

phds = [
  861, #2000
  830, #2001
  809, #2002
  867, #2003
  948, #2004
  1129, #2005
  1453, #2006
  1656, #2007
  1787, #2008
  1611, #2009
]

revenue = [
  1196000000, #2000
  1176000000, #2001
  1269000000, #2002
  1240000000, #2003
  1307000000, #2004
  1435000000, #2005
  1601000000, #2006
  1654000000, #2007
  1803000000, #2008
  1734000000, #2009
]

ax1 = plt.gca()
ax1.set_title("Comp Sci PhD vs Arcade Revenue")
line1, = ax1.plot(years, phds)
ax2 = ax1.twinx()
line2, = ax2.plot(years, revenue, c='r')
ax1.set_ylabel("Math PhDs awarded")
ax1.set_xlabel("Years")
ax2.set_ylabel("Revenue by arcades")
ax2.legend((line1, line2), ("math PhDs awarded", "revenue by arcades"))
```

## Part 2

1.   Create a pair plot of the iris dataset using Matplotlib. 
2.   Load data with Scikit-learn
3.   Label axes
4.   Diagonals need to contain histograms
5.   Distinguish different species by color
6.   Add a legend for the species 




```python
from sklearn.datasets import load_iris

iris = load_iris()
fig, ax = plt.subplots(4, 4, figsize=(15,15))
colors= ['turquoise', 'salmon', 'lime']

## Column 1
ax[0,0].hist(iris.data[iris.target==0, 0], label=iris.target_names[0], color=colors[0])
ax[0,0].set_ylabel(iris.feature_names[0])  
ax[0,0].hist(iris.data[iris.target==1, 0], label=iris.target_names[1], color=colors[1])
ax[0,0].hist(iris.data[iris.target==2, 0], label=iris.target_names[2], color=colors[2])
ax[0,0].legend(title="species")
ax[1,0].scatter(iris.data[:, 0], iris.data[:, 1], c=iris.target)
ax[1,0].set_ylabel(iris.feature_names[1])  
ax[2,0].scatter(iris.data[:, 0], iris.data[:, 2], c=iris.target)
ax[2,0].set_ylabel(iris.feature_names[2])  
ax[3,0].scatter(iris.data[:, 0], iris.data[:, 3], c=iris.target)
ax[3,0].set_ylabel(iris.feature_names[3])  
ax[3,0].set_xlabel(iris.feature_names[0])  

# Column 2
ax[0,1].scatter(iris.data[:, 1], iris.data[:, 0], c=iris.target)
ax[1,1].hist(iris.data[iris.target==0, 1], label=iris.target_names[0], color=colors[0])
ax[1,1].hist(iris.data[iris.target==1, 1], label=iris.target_names[1], color=colors[1])
ax[1,1].hist(iris.data[iris.target==2, 1], label=iris.target_names[2], color=colors[2])
ax[1,1].legend(title="species")
ax[2,1].scatter(iris.data[:, 1], iris.data[:, 2], c=iris.target)
ax[3,1].scatter(iris.data[:, 1], iris.data[:, 3], c=iris.target)
ax[3,1].set_xlabel(iris.feature_names[1])  

# Column 3
ax[0,2].scatter(iris.data[:, 2], iris.data[:, 0], c=iris.target)
ax[1,2].scatter(iris.data[:, 2], iris.data[:, 1], c=iris.target)
ax[2,2].hist(iris.data[iris.target==0, 2], label=iris.target_names[0], color=colors[0])
ax[2,2].hist(iris.data[iris.target==1, 2], label=iris.target_names[1], color=colors[1])
ax[2,2].hist(iris.data[iris.target==2, 2], label=iris.target_names[2], color=colors[2])
ax[2,2].legend(title="species")
ax[3,2].scatter(iris.data[:, 2], iris.data[:, 3], c=iris.target)
ax[3,2].set_xlabel(iris.feature_names[2])  

# Column 4
ax[0,3].scatter(iris.data[:, 3], iris.data[:, 0], c=iris.target)
ax[1,3].scatter(iris.data[:, 3], iris.data[:, 1], c=iris.target)
ax[2,3].scatter(iris.data[:, 3], iris.data[:, 2], c=iris.target)
ax[3,3].hist(iris.data[iris.target==0, 3], label=iris.target_names[0], color=colors[0])
ax[3,3].hist(iris.data[iris.target==1, 3], label=iris.target_names[1], color=colors[1])  
ax[3,3].hist(iris.data[iris.target==2, 3], label=iris.target_names[2], color=colors[2])
ax[3,3].set_xlabel(iris.feature_names[3])  
ax[3,3].legend(title="species")
```

## Part 3


#### Recreating figure 18.1 from "Fundamentals of Data Visualization" https://serialmentor.com/dataviz/overlapping-points.html

```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/pratoolbharti/NIU/master/Applied_Machine_Learning_S20/Homework-1/mpg.csv")

# FWD
displ_f = []
mpg_f = []

# RWD
displ_r = []
mpg_r = []

# 4WD
displ_4 = []
mpg_4 = []

count = 0
for item in df['drv']:
  if item == "f":
    displ_f.append(df['displ'][count])
    mpg_f.append(df['hwy'][count])
  elif item == "r":
    displ_r.append(df['displ'][count])
    mpg_r.append(df['hwy'][count])
  elif item == "4":
    displ_4.append(df['displ'][count])
    mpg_4.append(df['hwy'][count])
  count += 1

# plotting, everything above was data gen & import
plt.figure(figsize=(16,9))
plt.scatter(displ_f, mpg_f, label="FWD", c="tan")
plt.scatter(displ_r, mpg_r, label="RWD", c="lightblue")
plt.scatter(displ_4, mpg_4, label="4WD", c="black")
plt.legend(title="drive train")
plt.xlabel("displacement (I)")
plt.ylabel("fuel economy (mpg)")
```

#### Recreating figure 18.2 With partial transparency to show overlapping points

```python
# added alpha to help make points transparent
plt.figure(figsize=(16,9))
plt.scatter(displ_f, mpg_f, label="FWD", c="tan", alpha=.3, s=75)
plt.scatter(displ_r, mpg_r, label="RWD", c="lightblue", alpha=.3, s=75)
plt.scatter(displ_4, mpg_4, label="4WD", c="black", alpha=.3, s=75)

plt.legend(title="drive train")
plt.xlabel("displacement (I)")
plt.ylabel("fuel economy (mpg)")
```

#### 18.3 - Added jitter function to offset each point, could've just imported seaborn but you'd have to write a custom function for matplotlib

```python
# Added jitter function that is random and doesn't vary too much
# importan to not skew the data too much

def rand_jitter(arr):
    npArray = np.array(arr)
    stdev = npArray.max()/100.
    return arr + np.random.randn(len(arr)) * stdev

plt.figure(figsize=(16,9))
plt.scatter(rand_jitter(displ_f), rand_jitter(mpg_f), label="FWD", c="tan", alpha=.3, s=75)
plt.scatter(rand_jitter(displ_r), rand_jitter(mpg_r), label="RWD", c="lightblue", alpha=.3, s=75)
plt.scatter(rand_jitter(displ_4), rand_jitter(mpg_4), label="4WD", c="black", alpha=.3, s=75)

plt.legend(title="drive train")
plt.xlabel("displacement (I)")
plt.ylabel("fuel economy (mpg)")
```

#### 18.4 - Added jitter but way too much

```python
# Here I just increased the std deviation by doubling it making the data way too
# spread out, to the point it's misleading
def rand_jitter(arr):
    npArray = np.array(arr)
    stdev = npArray.max()/100.
    return arr + np.random.randn(len(arr)) * (stdev * 2)

plt.figure(figsize=(16,9))
plt.scatter(rand_jitter(displ_f), rand_jitter(mpg_f), label="FWD", c="tan", alpha=.3, s=75)
plt.scatter(rand_jitter(displ_r), rand_jitter(mpg_r), label="RWD", c="lightblue", alpha=.3, s=75)
plt.scatter(rand_jitter(displ_4), rand_jitter(mpg_4), label="4WD", c="black", alpha=.3, s=75)

plt.legend(title="drive train")
plt.xlabel("displacement (I)")
plt.ylabel("fuel economy (mpg)")
```
